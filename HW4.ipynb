```python
def crawl_gnn_pages(start_page=1, end_page=3):
    base_url = "https://gnn.gamer.com.tw/index.php?k="
    results = []

    for page in range(start_page, end_page + 1):
        url = f"{base_url}&page={page}"
        print(f"ğŸ“„ æ­£åœ¨æŠ“å–ç¬¬ {page} é : {url}")

        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                          "AppleWebKit/537.36 (KHTML, like Gecko) "
                          "Chrome/120.0 Safari/537.36"
        }

        res = requests.get(url, headers=headers)
        res.encoding = "utf-8"
        soup = BeautifulSoup(res.text, "html.parser")

        # æ¯ç¯‡æ–‡ç« åœ¨ class="GN-lbox2B" å…§
        articles = soup.select("div.GN-lbox2B")

        for article in articles:
            title_tag = article.select_one("h1 a")
            date_tag = article.select_one("div.GN-lbox2C span")
            author_tag = article.select_one("div.GN-lbox2C b")

            if title_tag:
                title = title_tag.text.strip()
                href = title_tag["href"]
                date = date_tag.text.strip() if date_tag else "æœªçŸ¥"
                author = author_tag.text.strip() if author_tag else "æœªçŸ¥"

                results.append({
                    "title": title,
                    "href": href,
                    "date": date,
                    "author": author
                })

    return results


# ä¸»ç¨‹å¼å€
news_list = crawl_gnn_pages(1, 3)

print(f"\nå…±æŠ“å– {len(news_list)} ç­†æ–°èï¼š\n")
for i, news in enumerate(news_list, 1):
    print(f"{i}. [{news['date']}] {news['title']} ({news['author']})")
    print(f"   é€£çµï¼š{news['href']}")
```

```python
import requests
from bs4 import BeautifulSoup
import time

def fetch_article_content(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/120.0 Safari/537.36"
    }
    try:
        res = requests.get(url, headers=headers, timeout=10)
        res.encoding = "utf-8"
        soup = BeautifulSoup(res.text, "html.parser")

        # GNN æ–‡ç« ä¸»é«”
        content = soup.select_one("div.GN-lbox3B")
        text = content.get_text(separator="\n", strip=True) if content else ""
        return text
    except Exception as e:
        print(f"âŒ æŠ“å–å¤±æ•—ï¼š{url} -> {e}")
        return ""
```

```python
import time
from urllib.parse import urljoin
import re
import numpy as np
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer

BASE = "https://gnn.gamer.com.tw"

# 1) è‹¥æ²’æœ‰ content æ¬„ä½ï¼Œå…ˆå»ºç«‹ç©ºæ¬„ä¸¦æŠ“å–æ–‡ç« å…§å®¹
if "content" not in df.columns:
    df["content"] = ""

# å° content ç‚ºç©ºçš„åˆ—æ‰å»æŠ“ï¼Œé¿å…é‡è¤‡è«‹æ±‚
need_fetch = ~df["content"].astype(str).str.strip().astype(bool)

if "href" not in df.columns:
    raise SystemExit("DataFrame ç¼ºå°‘ href æ¬„ä½ï¼Œè«‹å…ˆç¢ºèªè¡¨æ ¼å…§å®¹ã€‚")

if need_fetch.any():
    print(f"ğŸ•¸ï¸ éœ€è¦æŠ“å–å…§å®¹çš„æ–‡ç« æ•¸ï¼š{need_fetch.sum()}")
    for i, (idx, row) in enumerate(df[need_fetch].iterrows(), start=1):
        title = row.get("title", "(ç„¡æ¨™é¡Œ)")
        raw_href = str(row.get("href", "")).strip()
        url = urljoin(BASE, raw_href)  # æŠŠ // æˆ– / ç›¸å°è·¯å¾‘è½‰æˆå®Œæ•´ç¶²å€
        print(f"ğŸ” æŠ“å–({i}/{need_fetch.sum()}): {title}\n   â†’ {url}")
        try:
            text = fetch_article_content(url)  # ç”¨ä½ å‰é¢å®šç¾©å¥½çš„æ“·å–å‡½å¼
        except Exception as e:
            print(f"âŒ æŠ“å–ä¾‹å¤–ï¼š{e}")
            text = ""
        df.at[idx, "content"] = text if isinstance(text, str) else ""
        time.sleep(0.5)

# 2) ä¸­æ–‡åˆ†è©å™¨
def jieba_tokenizer(text: str):
    return list(jieba.cut(text or ""))

# 3) åªå°éç©ºå…§å®¹åš TF-IDF
nonempty_mask = df["content"].astype(str).str.strip().astype(bool)
valid_idx = df.index[nonempty_mask].tolist()
corpus = df.loc[valid_idx, "content"].tolist()

# å…ˆæº–å‚™è¼¸å‡ºæ¬„ä½ï¼ˆå…¨éƒ¨å…ˆæ¸…ç©ºï¼‰
df["top_keywords"] = "æŠ“å–å¤±æ•—æˆ–ç„¡å…§å®¹"
for i in range(1, 11):
    df[f"kw{i}"] = ""
    df[f"score{i}"] = ""

if len(corpus) == 0:
    print("â— æ²’æœ‰å¯ä¾›åˆ†æçš„å…§å®¹ã€‚")
else:
    vectorizer = TfidfVectorizer(
        tokenizer=jieba_tokenizer,
        stop_words=["çš„","æ˜¯","äº†","åœ¨","å’Œ","ä¹Ÿ","æœ‰","èˆ‡","ç‚º","ä¸Š","æ–¼","ä¸­","åŠ"],
        max_df=0.95, min_df=1
    )
    tfidf_matrix = vectorizer.fit_transform(corpus)
    feature_names = vectorizer.get_feature_names_out()

    # åªä¿ç•™ä¸­æ–‡è©
    CN_WORD = re.compile(r'^[\u4e00-\u9fff]+$')

    for local_i, global_i in enumerate(valid_idx):
        row_vec = tfidf_matrix[local_i].toarray()[0]
        if not np.any(row_vec):
            df.at[global_i, "top_keywords"] = "ï¼ˆç„¡æœ‰æ•ˆè©å½™ï¼‰"
            continue

        order = row_vec.argsort()[::-1]  # ç”±å¤§åˆ°å°æ’åºç´¢å¼•
        ranked = [(feature_names[j], float(row_vec[j])) for j in order if row_vec[j] > 0]

        # éæ¿¾æˆç´”ä¸­æ–‡ï¼Œå†å–å‰ 10
        ranked_cn = [(w, s) for (w, s) in ranked if CN_WORD.match(w)]
        top10 = ranked_cn[:10]

        if not top10:
            df.at[global_i, "top_keywords"] = "ï¼ˆç„¡ä¸­æ–‡é—œéµå­—ï¼‰"
            continue

        df.at[global_i, "top_keywords"] = ", ".join([f"{w}({s:.4f})" for w, s in top10])
        for k, (w, s) in enumerate(top10, start=1):
            df.at[global_i, f"kw{k}"] = w
            df.at[global_i, f"score{k}"] = f"{s:.6f}"

# 4) ä½ æœ€å¾Œæƒ³ä¿ç•™ï¼å›å¯«çš„æ¬„ä½
cols = ["title","href","date","author","top_keywords"] + \
       [f"kw{i}" for i in range(1,11)] + [f"score{i}" for i in range(1,11)]
df = df[[c for c in cols if c in df.columns]]  # ä»¥é˜²æœ‰æ¬„ä½ä¸å­˜åœ¨

print("\nâœ… å·²å®Œæˆï¼šæŠ“å–ç¼ºæ¼å…§å®¹ + ä¸­æ–‡ TF-IDFï¼ˆå«åˆ†æ•¸èˆ‡ kw1..kw10ï¼‰")
df.head()
```

```python
import os
import google.generativeai as genai
os.environ["GOOGLE_API_KEY"]="AIzaSyA3ncs4c7FIMX0Wupc-zdCKapMUwDLoBO4"
GEMINI_MODEL = "gemini-2.5-pro"  # å¯æ”¹æˆä½ åå¥½çš„å‹è™Ÿ

def _build_corpus_for_llm(df, max_chars=8000):
    """æŠŠ df æ•´ç†çµ¦ LLMï¼›æ¯ç¯‡å–ï¼šæ¨™é¡Œ + é—œéµå­— + å…§æ–‡å‰ 1000 å­—ã€‚"""
    blocks, total = [], 0
    for _, r in df.iterrows():
        title = str(r.get("title", ""))[:80]
        kws = str(r.get("top_keywords", ""))  # è‹¥ä½ ç”¨ kw1..kw10ï¼Œä¹Ÿå¯è‡ªè¡Œçµ„åˆ
        text = str(r.get("content", ""))[:1000]
        chunk = f"ã€æ¨™é¡Œã€‘{title}\nã€é—œéµå­—ã€‘{kws}\nã€å…§æ–‡æ‘˜éŒ„ã€‘{text}\n"
        if total + len(chunk) > max_chars:
            break
        blocks.append(chunk)
        total += len(chunk)
    return "\n---\n".join(blocks)

def gemini_insights_and_summary(df):
    """è¼¸å…¥ä½ ç¾æœ‰çš„ dfï¼Œå›å‚³ (markdownæ–‡å­—)ã€‚"""
    api_key = os.environ.get("GOOGLE_API_KEY", "").strip()
    if not api_key:
        return "ï¼ˆæœªè¨­å®š GOOGLE_API_KEYï¼Œè«‹å…ˆè¨­å®šç’°å¢ƒè®Šæ•¸ï¼‰"
    genai.configure(api_key=api_key)

    corpus_text = _build_corpus_for_llm(df, max_chars=8000)
    if not corpus_text.strip():
        return "ï¼ˆæ²’æœ‰å¯ç”¨å…§å®¹å¯ä¾›æ‘˜è¦ï¼‰"

    prompt = f"""
ä½ æ˜¯ä¸­æ–‡è³‡æ–™åˆ†æåŠ©ç†ã€‚ä»¥ä¸‹æ˜¯å¤šç¯‡ GNN æ–°èçš„å…§å®¹æ‘˜éŒ„èˆ‡æ¯ç¯‡çš„ TF-IDF é—œéµè©ã€‚è«‹ç”¢å‡ºï¼š
1) **5 å¥**ç²¾ç…‰ã€Œæ´å¯Ÿæ‘˜è¦ã€ï¼Œç”¨æ¢åˆ—é»ï¼ˆæ¯é»ä¸è¶…é 40 å­—ï¼‰ã€‚
2) **120 å­—**å·¦å³çš„ã€Œç¸½çµçµè«–ã€ï¼ˆå–®æ®µè½ï¼Œç´„ 100~140 å­—ï¼‰ã€‚

èšç„¦ï¼šå…±åŒè¶¨å‹¢ã€ç”¨æˆ¶/å¸‚å ´æ„æ¶µã€å¯èƒ½è¡Œå‹•å»ºè­°ï¼›é¿å…é‡è¤‡æ¨™é¡Œï¼Œä¿æŒå®¢è§€ã€‚

=== å…§å®¹é–‹å§‹ ===
{corpus_text}
=== å…§å®¹çµæŸ ===
"""
    model = genai.GenerativeModel(GEMINI_MODEL)
    resp = model.generate_content(prompt)
    return resp.text.strip() if resp and resp.text else "ï¼ˆæœªå–å¾—æ¨¡å‹å›æ‡‰ï¼‰"
```

```python
# !pip install -q gradio
import gradio as gr
import pandas as pd

def _present_df_for_ui(df):
    """åªæŒ‘ä½ æƒ³é¡¯ç¤ºåœ¨è¡¨æ ¼çš„æ¬„ä½"""
    show_cols = ["title","date","author","kw1","kw2","kw3","kw4","kw5","kw6","kw7","kw8","kw9","kw10"]
    show_cols = [c for c in show_cols if c in df.columns]
    if not show_cols:
        # è‹¥ä½ æ²’æœ‰ kw1..kw10ï¼Œå°±é€€å›é¡¯ç¤º title + top_keywords
        show_cols = [c for c in ["title","date","author","top_keywords"] if c in df.columns]
    return df[show_cols].copy()

def _run_gemini(current_df):
    # Gradio æœƒæŠŠ Dataframe å‚³å› list of dictï¼Œè½‰å› DataFrame
    if isinstance(current_df, list):
        current_df = pd.DataFrame(current_df)
    return gemini_insights_and_summary(current_df)

with gr.Blocks(title="TF-IDF â†’ Gemini æ´å¯Ÿï¼ˆæœ€å° UIï¼‰") as demo:
    gr.Markdown("### ç›®å‰ DataFrameï¼ˆè«‹å…ˆåœ¨å¤–éƒ¨ç¨‹å¼å®Œæˆ content & TF-IDF å†é–‹ UIï¼‰")

    # ç”¨ä½ è¨˜æ†¶é«”ä¸­çš„ df åšé è¨­é¡¯ç¤º
    df_default = _present_df_for_ui(df)
    df_in = gr.Dataframe(value=df_default, label="(å¯ç·¨ä¿®å¾Œå†æ‘˜è¦)", wrap=True)

    btn = gr.Button("ç”Ÿæˆï¼š5 å¥æ´å¯Ÿ + 120 å­—çµè«–ï¼ˆGeminiï¼‰")
    md = gr.Markdown()

    btn.click(fn=_run_gemini, inputs=[df_in], outputs=[md])

demo.launch()  # ä½ å¯åœ¨éœ€è¦æ™‚å†æ‰‹å‹•å•Ÿå‹•
```
